### Count both the total video count and the unique video count for each country

```python
query = """
    SELECT 
        country, 
        COUNT(video_id) AS total_video_count, 
        COUNT(DISTINCT video_id) AS unique_video_count
    FROM Videos
    GROUP BY country
    ORDER BY total_video_count DESC
"""

cur.execute(query)

results = cur.fetchall()

print("Country             | Total Video Count | Unique Video Count")
print("--------------------------------------------------------------")
for row in results:
    print(f"{row[0]:<20} | {row[1]:<17} | {row[2]}")
#US has the highest total video count,it has a smaller number of unique videos, indicating a higher incidence of duplicate videos compared to other countries
#Japan has the lowest total video count and unique videos count,there may be a higher prevalence of duplicate videos or a smaller diversity of content compared to countries with higher unique video counts.
```

### Most view_count category

```python
query2 = """
SELECT category, SUM(view_count) AS total_view_count
FROM videos
GROUP BY category
ORDER BY total_view_count DESC
LIMIT 1;
"""

cur.execute(query2)

result = cur.fetchone()

if result:
    print("Category with the most view count:")
    print("Category:", result[0])
    print("Total View Count:", result[1])
else:
    print("No data found.")
```

### Top five categories within each country based on total view count

```python
query3 = """
SELECT country, category, total_view_count, category_rank
FROM (
    SELECT country, category, SUM(view_count) AS total_view_count,
           RANK() OVER (PARTITION BY country ORDER BY SUM(view_count) DESC) AS category_rank
    FROM videos
    GROUP BY country, category
) AS ranked_categories
ORDER BY country, category_rank;
"""

cur.execute(query3)

results_by_country = cur.fetchall()

print("Country      | Rank | Category            | Total View Count")
print("------------------------------------------------------------")
for row in results_by_country:
    if row[3] <= 5:  # Check if category rank is within top five
        print(f"{row[0]:<12} | {row[3]:<4} | {row[1]:<20} | {row[2]}")
        
#1st Music for Brazil and Japan, Entertainment for Great Britain and US
```

### Which category by country get the most like and dislike

```python
query4 = """
WITH ranked_categories AS (
    SELECT category,
           country,
           SUM(likes) AS total_likes,
           SUM(dislikes) AS total_dislikes,
           ROW_NUMBER() OVER (PARTITION BY country ORDER BY SUM(likes) DESC, SUM(dislikes) DESC) AS rank_within_country
    FROM videos
    GROUP BY category, country
)
SELECT *
FROM ranked_categories
WHERE rank_within_country <= 10
ORDER BY country, rank_within_country;
"""

cur.execute(query4)

results = cur.fetchall()

print("Country   | Rank | Category            | Total Likes | Total Dislikes")
print("----------------------------------------------------------------------")
for row in results:
    print(f"{row[1]:<9} | {row[4]:<4} | {row[0]:<20} | {row[2]:<12} | {row[3]}")
    
#The top categories in all countries are dominated by Music, Entertainment, Gaming, People & Blogs.
#Music and Entertainment have significantly higher likes and dislikes, meaning a large audience and evoke strong reactions from viewers.
```

### Top 5 categories within each country based on the count of videos

```python
query5 = """
WITH ranked_categories AS (
    SELECT category,
           country,
           COUNT(*) AS count,
           COUNT(DISTINCT view_count) AS nunique_view_count,
           ROW_NUMBER() OVER (PARTITION BY country ORDER BY COUNT(*) DESC) AS rank_within_country
    FROM videos
    GROUP BY category, country
)
SELECT *
FROM ranked_categories
WHERE rank_within_country <= 5
ORDER BY country, rank_within_country;
"""

cur.execute(query5)

results = cur.fetchall()

print("Country   | Rank | Category            | count | nunique_view_count")
print("-------------------------------------------------------------------")
for row in results:
    print(f"{row[1]:<9} | {row[4]:<4} | {row[0]:<20} | {row[2]:<5} | {row[3]}")
    
#Entertainment seems to be a popular category across all countries, followed by Gaming, Music, Sports, and People & Blogs. T
#he higher count of unique view counts in Entertainment and Music categories suggests that these categories might have a more engaged audience or more viral content compared to others.
```

### Count the top 10 duplicate video_id entries along with their corresponding categories

```python
query6 = """
WITH duplicate_video_ids AS (
    SELECT video_id, category, COUNT(*) AS duplicate_count,
           ROW_NUMBER() OVER (ORDER BY COUNT(*) DESC) AS row_num
    FROM videos
    GROUP BY video_id, category
    HAVING COUNT(*) > 1
)
SELECT video_id, category, duplicate_count
FROM duplicate_video_ids
WHERE row_num <= 10;
"""

cur.execute(query6)

results = cur.fetchall()

print("Top 10 Duplicate Video IDs with Category")
print("----------------------------")
for row in results:
    print(f"Video ID: {row[0]}, Category: {row[1]}, Duplicate Count: {row[2]}")
    
#High duplicate frequency includes Entertainment, Music, Gaming, and Sports
```

### Count the last 10 duplicate video_id entries along with their corresponding categories

```python
query7 = """
WITH duplicate_video_ids AS (
    SELECT video_id, category, COUNT(*) AS duplicate_count
    FROM videos
    GROUP BY video_id, category
    HAVING COUNT(*) > 1
)
SELECT video_id, category, duplicate_count
FROM duplicate_video_ids
ORDER BY duplicate_count DESC
LIMIT 10 OFFSET (
    SELECT COUNT(*) - 10 FROM duplicate_video_ids
);
"""

cur.execute(query7)

results = cur.fetchall()

print("Last 10 Duplicate Video IDs with Category")
print("----------------------------")
for row in results:
    print(f"Video ID: {row[0]}, Category: {row[1]}, Duplicate Count: {row[2]}")
    
#Low duplicate frequency includes People & Blogs, Music, Sports, and Comedy
```

###  Top 5 categories within each country based on the tags

```python
#Split the tags and count the occurrences of each category within each country
query8 = """
WITH tag_categories AS (
    SELECT video_id, UNNEST(string_to_array(tags, ',')) AS tag
    FROM videos
)
SELECT country, tag AS category, COUNT(*) AS tag_count
FROM tag_categories
JOIN videos ON tag_categories.video_id = videos.video_id
GROUP BY country, tag
ORDER BY country, tag_count DESC
"""

cur.execute(query8)

results = cur.fetchall()

print("Top 5 Categories within Each Country Based on Tags")
print("--------------------------------------------------")
current_country = None
country_category_count = 0
for row in results:
    if current_country != row[0]:
        current_country = row[0]
        country_category_count = 0
        print(f"\n{current_country}:")
    if country_category_count < 5:
        print(f"{row[1]}: {row[2]}")
        country_category_count += 1
        
#music labels (HYBE) and gaming (genshin impact) appear to be popular across multiple countries. 
#Country-specific interests, challenges (Desafio in Brazil, sidemen in Great Britain) and Japanese content creators (hajime in Japan) are also reflected in the tags.
```
